```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('Mall_Customers.csv')
```

```python
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
n_clusters = 3
```

```python
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
data['Cluster'] = kmeans.fit_predict(X)

# Visualize the clusters
plt.figure(figsize=(12, 6))

# Scatter plot
plt.subplot(1, 2, 1)
sns.scatterplot(data=data, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='viridis', legend='full')

# Bar plot
plt.subplot(1, 2, 2)
#sns.barplot(data=data, x='Cluster', y='Age', estimator='count', palette='viridis')
sns.barplot(data=data, x='Cluster', y='Age', estimator=len, palette='viridis', hue='Cluster', dodge=False, errorbar=None)
plt.legend().set_visible(False)
plt.show()
```
